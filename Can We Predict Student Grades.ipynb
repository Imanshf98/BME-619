```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# shuffle module is useful to do random permutations of the collections on arrays.
from sklearn.utils import shuffle
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/student-mat.csv")
data.head(5)
```

```python
# In this dataset the G1 represents the grades of the first period, G2 represents the grades of the second period,
# and G3 represents the final grades.
# I split the dataset into 30% testing and 70% training
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)
accuracy = lr.score(xtest, ytest)
print(accuracy)
```

```python
predictions = lr.predict(xtest)
for i in range(len(predictions)):
    print(predictions[x], xtest[x], [ytest[x]])
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, xtrain, ytrain, scoring='r2', cv=5)
scores   
```
